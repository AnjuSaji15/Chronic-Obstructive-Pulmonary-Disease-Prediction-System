```python
from tensorflow.keras.models import load_model
import numpy as np
import pandas as pd
import joblib

cnn_model = load_model("cnn_model.h5")
ann_model = load_model("ann_model.h5")
scaler = joblib.load("scaler.pkl")
```

```python
from IPython.display import display
import ipywidgets as widgets
from PIL import Image
import io

uploader = widgets.FileUpload(accept='image/*', multiple=False)
display(uploader)
```

```python
if uploader.value:
    file = uploader.value[0]
    img_bytes = file['content']

    img = Image.open(io.BytesIO(img_bytes)).convert("RGB")
    img = img.resize((128,128))

    img = np.array(img) / 255.0
    img = np.expand_dims(img, axis=0)

    cnn_pred = cnn_model.predict(img)
    print("CNN:", cnn_pred)
```

```python
print("\n--- Enter Patient Details ---")

age = float(input("Age (years): "))
sex = int(input("Sex (0 = Female, 1 = Male): "))
height = float(input("Height (cm): "))
weight = float(input("Weight (kg): "))
bmi = float(input("BMI: "))
smoking = int(input("Smoking_Status (0 = No, 1 = Yes): "))
fev1 = float(input("FEV1_L (liters): "))
fvc = float(input("FVC_L (liters): "))
```

```python
import pandas as pd

patient_data = pd.DataFrame([[
    age,
    sex,
    height,
    weight,
    bmi,
    smoking,
    fev1,
    fvc
]], columns=[
    'Age',
    'Sex',
    'Height_cm',
    'Weight_kg',
    'BMI',
    'Smoking_Status',
    'FEV1_L',
    'FVC_L'
])
```

```python
print(scaler.feature_names_in_)
```

```python
patient_data = scaler.transform(patient_data)

ann_pred = ann_model.predict(patient_data)
print("ANN:", ann_pred)
```
